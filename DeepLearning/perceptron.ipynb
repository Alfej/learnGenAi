```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import Perceptron
from mlxtend.plotting import plot_decision_regions
```

```python
# Define the number of rows you want to generate
num_rows = 100

# Generate random data for each column
cgpa = np.random.uniform(6, 9, num_rows)
resume_score = np.random.uniform(5, 10, num_rows)
data = {
    'resume_score': resume_score,
    'CGPA': cgpa
}
df = pd.DataFrame(data)

# Apply the condition: 2*CGPA + 3*resume_score > 7.5
df['Placed'] = (35.76717307 * df['CGPA'] - 38.53268781 * df['resume_score']) > 0
df['Placed'] = df['Placed'].apply(lambda x: 1 if x == True else 0).to_numpy()

# Convert boolean to 'Yes' or 'No'
# df['Placed'] = df['Placed'].apply(lambda x: 'Yes' if x else 'No')

# Display the DataFrame
print(df)
```

```python
sns.scatterplot(data=df, x="CGPA", y="resume_score", hue="Placed")
```

```python
x = df.iloc[:,0:2]
y = df.iloc[:,-1]  
```

```python
p = Perceptron(max_iter=5000,alpha=0.1)
```

```python
from mlxtend.plotting import plot_decision_regions
import numpy as np

# Extract only numeric columns (resume_score, CGPA) from the DataFrame
x = df[['resume_score', 'CGPA']].to_numpy()  # Convert DataFrame to NumPy array

# Convert 'Placed' column ('Yes'/'No') to integers (1/0)
y = df['Placed'].to_numpy()  # Convert to NumPy array

# Now plot the decision regions using your classifier `p`
plot_decision_regions(x, y,clf=p)
```

```python
predicted = p.predict(x)
```
